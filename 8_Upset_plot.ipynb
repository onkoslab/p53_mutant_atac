## Upset plot for Figure1 E&F

```r
library('UpSetR')
```

```r
val='brca'
df<- read.csv(paste(paste('Results/',val,sep=''),'_upset.tsv',sep=''),sep='\t', header=TRUE)
colnames(df)=c('name','region','P53 RE','Mutant Chip Sites','None',"Copy Number Gain" ,"Copy Number Loss" )

par(family='serif',font.axis=2)
slam_colours = c('#66c2a5','#fc8d62','#8da0cb','#e78ac3','#a6d854') 

cat = c("None" ,"P53 RE","Mutant Chip Sites","Copy Number Gain" ,"Copy Number Loss" )

#svg(paste(paste('plots/',val,sep=''),'_upset.svg',sep='')) 
upset(
  df,
  sets = cat,
  keep.order = TRUE,
  sets.bar.color = slam_colours,
  text.scale = c(1.3, 1.3, 1.3, 1, 1.5, 1),
  mainbar.y.label = paste(val," Number of peaks",sep=''),
  sets.x.label = "Feature Count",
  main.bar.color='#c51b8a',
  query.legend = "bottom",
  ylim(0,100),
  queries = list(
    list(
      query = elements,
      params = list("region", "Loss"),
      active = T,
      color='#3182bd',
      query.name = "Loss"
    )
  )
)
```
